## The objective of this exercise is to understand the Metropolis-Hastings algorithm, a Markov chain Monte Carlo (MCMC) method for sampling.

### Plot for Sigma (low=0.5)


```python
import math
import numpy as np
import matplotlib.pyplot as plot
sd=0.5
N=1500
plot_pt=[]
iteration=[]

#proposal distribution function
def prop(x,mean):
    return (1/(math.sqrt(2*math.pi*sd**2)))*(math.e**(-((x-mean)**2)/(2*sd**2)))

#target distribution function
def target(x):
    return (np.e**(-(x**4)))*(2+np.sin(5*x)+np.sin(-(2*x*x)))

#MH algorithm
 
def metropolis_hasting(N):
    x_o = -1
    p= target(x_o)
    
    for i in range(N):
        x_candidate=np.random.normal(x_o, sd, size= None)
        p_next=target(x_candidate)
        alpha=min(1,(prop(x_o,x_candidate)*p_next)/(prop(x_candidate,x_o)*p))
        u = np.random.uniform(0,1)
        if u < alpha:
            p = p_next
            x_o = x_candidate
        plot_pt.append(x_o)
        iteration.append(i)
        
def plot_graphs():
    rng = np.arange(-2.5,2.5,.1)
     #Histogram for generated sample 
    plot.hist(plot_pt, bins=1500,density=True)
    plot.title('sigma = %s'%sd)
    plot.xlabel('x')
    plot.ylabel('Probability Density P(x)')
    #Actual distribution 
    plot.plot(rng,target(rng),color='g',linewidth=3)   
    plot.show()
    
    #Generated Sample vs Iteration
    plot.plot(iteration,plot_pt)
    plot.title('Generated Sample vs Iteration')
    plot.xlabel('Iteration Number')
    plot.ylabel('Generated Sample')
    plot.show()
    
    
metropolis_hasting(N)
plot_graphs()
```
